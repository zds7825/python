####요구사항 1. 생성자에 count 값과 step 값을 인자로 받을 수 있다.
####count: Counter 인스턴스가 지니는 초기 정수 값

####>>> c.incr()

####요구사항 4. Counter 클래스 내에 관련 메소드를 추가하여 인스턴스 객체를 직접 호출(call)할 수 있도록 하시오. 인스턴스 객체를 직접 호출했을 때에 내부적으로 incr() 메소드를 호출하는 방법으로 구현하시오.
####>>> c()

####요구사항 5. 다음과 같은 두 개의 산술 연산 (+, -)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.
####>>> c = c + 5 
####>>> d = d - 5 
####>>> print c 
####[Count (step: 1)] 17 
####>>> print d 
####[Count (step: 2)] 9 
####요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.
####>>> print c > 10 
####True 
####>>> print d > 10 
####False 
####>>> print c < 10 
####False 
####>>> print d < 10 
####True 
####>>> print c == 17 
####True 
####>>> print d != 9 
####False 

```python
class Counter:
    def __init__(self, count ,step =1):
        self.count = count
        self.step = step
        
    def __str__(self):
        return '[Count(step : '+ str(self.step)+ ')] ' + str(self.count)
    
    def __call__(self):
        self.incr()
        
    def incr(self):
        self.count = self.count + self.step
        
    def __add__(self, n):
        return Counter(self.count + n, self.step)
        
    def __sub__(self, n):
        return Counter(self.count - n, self.step)
        
    def __cmp__(self, n):
        return self.count - n
    
    
```

```python
c = Counter(10)
d = Counter(10, 2)
print c
print d
```

```python
c.incr()
d.incr()
print c
print d
```

```python
c()
d()
print c
print d
```

```python
c = c+5
d = d-5
print c
print d
```

```python
print c>10
print d>10
print c<10
print d<10
print c==17
print d!=9
```

####__init__을 통하여 클래스 초기화의 조건을 만들어 주었다.  __str__을 통하여 함수가 print되었을 경우 출력되는 내용을 정해주었다.  각 연산자들을 재정의 해주었으며, __cmp__재정의를 통하여 클래스 인스턴스와 숫자와의 연산이 가능하게 하였다.

##문제 4
####다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.
####MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t
```

####__init__은 클래스의 생성자로써 클래스가 선언이 될 때 바로 선언이 된다.  생성자는 list를 매개변수로 받음으로 써 리스트의 요소들을 클래스 내부에 리스트를 만들어 넣는다. 그리고 eliminate_duplicate()를 실행한다. __str__은 클래스를 print로 출력할 때 선언되는 함수이다. 이 코드에서는 클래스 내부의 리스트를 문자열 형태로 바꾸어 출력하였다. dliminate_duplicate()는 매개변수로 리스트를 받은 후, 중복되는 요소가 있으면 제거한다.

##문제 5
####4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오
####정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.
####요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.
####>>> u = s | t 
####>>> print u 
####MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9} 
####요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.
####>>> u = s & t 
####>>> print u 
####MySet: {2, 3} 
####요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.
####>>> s = MySet([1, 2, 3]) 
####>>> t = MySet([3, 4, 5]) 
####>>> u = s - t 
####>>> print u 
####MySet: {1} 

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
            
    def __or__(self, l):
        l2 = []
        for e in l:
            l2.append(e)
        for e in self:
            if e not in l2:
                l2.append(e)
                l2.sort()
        return MySet(l2)
    
    def __and__(self, l):
        l2 = []
        for e in l:
            if e in self:
                l2.append(e)
        return MySet(l2)
        
    def __sub__(self, l):
        l2 = []
        for e in self:
            l2.append(e)
        for e in l2:
            if e in l:
                l2.remove(e)
        return MySet(l2)
        

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t
    
    u = s|t
    print u
    
    u = s&t
    print u
    
    s = MySet([1, 2, 3])
    t = MySet([3, 4, 5])
    u = s - t
    print u
```

####클래스는 list를 상속받았기 때문에 __init__은 클래스를 객체에 넣어 만들 경우 변수를 리스트로 받아 자신을 리스트에 모두 넣어주었다. 그리고 각 집합과 관련된 함수들을 재정의 해주었다.

##오일러 17
####1부터 5까지의 숫자를 영어로 쓰면 one, two, three, four, five 이고, 각 단어의 길이를 더하면 3 + 3 + 5 + 4 + 4 = 19 이므로 사용된 글자는 모두 19개입니다.
####1부터 1,000까지 영어로 썼을 때는 모두 몇 개의 글자를 사용해야 할까요?
####참고: 빈 칸이나 하이픈('-')은 셈에서 제외하며, 단어 사이의 and 는 셈에 넣습니다. 예를 들어 342를 영어로 쓰면 three hundred and forty-two 가 되어서 23 글자, 115 = one hundred and fifteen 의 경우에는 20 글자가 됩니다.

```python
def numberCount(x):
    if x==1000:
        return 11
    elif x>=100:
        return count3(x)

    elif x >=20:
        return count2(x)

    else:
        if x == 0:
            return 0
        else:
            return count1(x)
        
def count1(x):
    if x == 0:
        return 0
    elif x ==1 or x == 2 or x == 6 or x == 10:
        return 3
    elif x == 4 or x == 5 or x == 9:
        return 4
    elif x == 3 or x == 7 or x == 8: 
        return 5
    elif x == 11 or x == 12:
        return 6
    elif x == 15 or x == 16:
        return 7
    elif x == 13 or x == 14 or x == 19 or x == 18:
        return 8
    elif x == 17:
        return 9
    
def count2(x):
    if x/10 == 2 or x/10 ==3 or x/10 ==5 or x/10 == 8 or x/10 ==4:
        if x/10 == 2 or x/10 == 3 or x/10 == 8:
            return 6 + count1(x%10)
        else:
            return 5 + count1(x%10)
    else:
        return count1(x/10)+ 2 + count1(x%10)
    
def count3(x):
    if x%100 == 0:
        if x%100 >=20:
            return count1(x/100) + 7 + count2(x%100)
        else:
            return count1(x/100) + 7 + count1(x%100)
    else:
        if x%100 >=20:
            return count1(x/100) + 10 + count2(x%100)
        else:
            return count1(x/100) + 10 + count1(x%100)
    

result = 0
for x in range(1, 1001):
    result = result + numberCount(x)
print result
```

####1000일때, 100이상일 경우, 20이상일 경우, 19미만일 경우를 나누어 숫자를 판단하는 함수를 만들어 준다. 4가지 함수를 모두 이용하여 알파벳 수를 출력한다.

####•1900년 1월 1일은 월요일이다.
####•4월, 6월, 9월, 11월은 30일까지 있고, 1월, 3월, 5월, 7월, 8월, 10월, 12월은 31일까지 있다.

##오일러 20
####n! 이라는 표기법은 n × (n − 1) × ... × 3 × 2 × 1을 뜻합니다.
####예를 들자면 10! = 10 × 9 × ... × 3 × 2 × 1 = 3628800 이 되는데,
####여기서 10!의 각 자리수를 더해 보면 3 + 6 + 2 + 8 + 8 + 0 + 0 = 27 입니다.
####100! 의 자리수를 모두 더하면 얼마입니까?


```python
def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n-1)
    
s = str(factorial(100))
result = 0
for x in range(0, len(s)):
    result = result + int(s[x])
print result
```

####100!의 값을 문자열로 만든 후, 각 자리를 인덱싱하여 모두 더하였다.

##오일러 21
####n의 약수들 중에서 자신을 제외한 것의 합을 d(n)으로 정의했을 때,
####서로 다른 두 정수 a, b에 대하여 d(a) = b 이고 d(b) = a 이면 
########a, b는 친화쌍이라 하고 a와 b를 각각 친화수(우애수)라고 합니다. 
####예를 들어 220의 약수는 자신을 제외하면 1, 2, 4, 5, 10, 11, 20, 22, 44, 55, 110 이므로 그 합은 d(220) = 284 입니다.
####또 284의 약수는 자신을 제외하면 1, 2, 4, 71, 142 이므로 d(284) = 220 입니다.
####따라서 220과 284는 친화쌍이 됩니다. 
####10000 이하의 친화수들을 모두 찾아서 그 합을 구하세요.


```python
def d(n):
    result = 0
    for x in range(1, n):
        if n%x == 0:
            result = result + x
    return result

result = 0
for n in range(1, 10001):
    if n == d(d(n)) and n != d(n):
        if d(n) <= 10000:
            result = result + n
print result
            
```

####함수 d()는 입력된 수의 약수들을 모두 더한 값을 반환해준다.  반복문을 통해 친화쌍을 구하여 더해주었지만, 약수를 찾아 더한 값이 다시 자신이 되는 것들을 빼주어야 한다.

```python
import urllib2
import string    #functuation을 불러들이기 위해



def FrequencyCount(url):
    source = urllib2.urlopen(url).read()
    for i in range(0, source.count('<!--')):
        if source.find("<!--") != -1:
            source = source.replace(source[source.find("<!--"):source.find("-->")+3:],"")

    length = len(source)
    s = ""
    n = 1
    for x in range(0, length):
        if n==1:
            if source[x] == '<':
                n = 0
                pass
            else:
                s = s + source[x]
        elif n==0:
            if source[x] == '>':
                n = 1
            else:
                pass
    L = s.split()

    import string
    punc = string.punctuation

    for x in range(0, len(L)):
        for y in punc:
            if L[x].find(y) != -1:
                L[x] = L[x].replace(y, "")
            
            

    count = 0
    for x in range(0, len(L)):
        if L[x] == "" or L[x] == " ":
            count = count + 1



    while(count >0):
        for x in range(0, len(L)):
            if L[x] == "":
                del L[x]
                count = count -1
                break

    word = [ 'a', 'and', 'are', 'as', 'at', 'be', 'but', 'by', 'for', 'if', 'in', 'into', 'is', 'it', 'no', 'not', 'of', 'on', 'or', 's', 'such', 't', 'that', 'the', 'their', 'then', 'there', 'these', 'they', 'this', 'to', 'was', 'will', 'with']



    count = 0
    for x in range(0, len(L)):
        if L[x] in word:
            count = count +1

    while(count >0):    #영어 불용어를 모두 제거하기 위해 사용
        for x in range(0, len(L)):
            if L[x] in word:
                del L[x]
                count = count -1
                break



    d = {}

    for x in range(0, len(L)):
        for y in range (0, len(L)):
            if L[x] == L[y]: 
                count = count + 1
        d[L[x]] = count
        count = 0
    
    return d
    




class WebWordsFrequency:
    l = []
    def __init__(self, *url):    #생성자
        for x in url:
            self.l.append(x)
            
    def addUrl(self, url):    #url을 추가
        self.l.append(url)
        
        
    def removeUrl(self, url):    #url을 삭제
        self.l.remove(url)
        
    
    def listUrls(self):     #클래스의 url모두 출력
        for x in self.l:
            print x
            
    def getWordsFrequency(self):
        total_d = {}
        for x in self.l:
            d = FrequencyCount(x)
            for x in d:
                if x in total_d:
                    total_d[x] = total_d[x] + d[x]
                else:
                    total_d[x] = d[x]
        
        
        if total_d == {}:
            return None
        else:
            return total_d
    
    def getMaxFrequencyWord(self):
        l2 = []
        d = self.getWordsFrequency()

        
        max_word = ""
        max_count = 0
        
        for x in d:
            if d[x] > max_count:
                max_count = d[x]
                max_word = x
        for x in d:
            if max_count == d[x]:
                l2.append(x)
        if l2 == []:
            return None
        else:
            return l2
        
        
            


```
