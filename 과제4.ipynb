##문제 3
####range() 함수와 유사한 frange() 함수를 다음 조건을 참고하여 만드시오.
####•1) frange()의 인자 구성은 다음과 같이 range와 동일하지만 각 인수들은 음수를 받지 않는다고 가정한다.
####range(stop)
####range(start, stop[, step])
####•2) frange() 함수의 인자에 대한 기본 시작(start) 값은 0.0이고, 기본 단계(step) 값은 0.1이다. 
####•3) frange 사용 예◾3-1) frange(0.5)◦[0.0, 0.1, 0.2, 0.3, 0.4]
####◾3-2) frange(1.0, 2.0)◦[1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9]
####◾3-3) frange(2.2, 4.0, 0.5)◦[2.2, 2.7, 3.2, 3.7]


```python
def frange(start, end = -1, step=0.1):
    l = []
    if end == -1:
        x = 0
        while x < start:
            l = l + [round(x, 1)]
            x = x + step
    else:
        x = start
        while x < end:
            l = l + [round(x, 1)]
            x = x + step
    return l
            

print frange(0.5)
print frange(1.0, 2.0)
print frange(2.2, 4.0, 0.5)
```

#### 이 문제에서 가장 중요한 부분은 매개변수에 기본 인수 값을 정의해주는 것이다. 기본적으로 음수의 값을 안받는다고 정의 했으므로, 변수를 1개만  받는 range(stop)을 구현할 경우 end = -1으로 기본 인수를 정의 하고 그 경우에는 start 매개변수가 end 역할을 하게 해준다. round를 사용하지 않았을 경우 긴 소수가 출력될 수 있으므로 반올림 해주었다.


##문제 4
####가변인수를 받는 함수 sum()을 다음과 같은 조건을 참고하여 구현하시오
####•sum() 사용 예 ◾sum()◦0
####◾sum(1, 2)◦3
####◾sum(1, 2, 3, 4, 5)◦15
####◾sum(1, 5, 7, 2, -10)◦5


```python
def sum(*arg):
    n = 0
    for x in arg:
        n = n + x
    return n

print sum()
print sum(1, 2)
print sum(1, 2, 3, 4, 5)
print sum(1, 5, 7, 2, -10)
```

####sum()의 매개변수는 가변인수를 받는다. 매개변수의 합을 나타내는 변수 n을 만들어 주며, 매개변수의 원소들을 하나씩 꺼내어 모두 n에 더해준다.

####여러 단어로 이루어진 문자열을 입력받아 각 단어의 첫글자로 이루어진 단어를 대문자로 출력하는 myinitial() 함수를 다음 조건을 참고하여 작성하시오.
####•1) 다음에 제시되는 함수들을 모두 이용해야 한다. ◾split
####◾map
####◾join
####•2) myinitial() 함수 사용 예◾myinitial("as soon as possible")◦ASAP

```python
def myinitial(s):
    l = s.split(" ")
    l = map(lambda a:a[0], l)
    s = "".join(l)
    return s.upper()

print myinitial('as soon as possible')
    
    
```

####myinitial()의 매개변수는 문자열을 입력받는다. l에 s를 ' '기준으로 리스트를 만들어 넣어준다. l 에 map()안에 lambda함수를 사용하여 l의 각 원소의 첫번째 부분만 추출하여 리스트를 만들어 넣어준다. s에 join을 사용하여 l을 모두 연결해 리스트로 넣어준다. 값을 반환할 시에 모두 대문자로 만들어 출력해준다.

##문제 6
####음이 아닌 정수 n를 입력받으면 n! (factorial)을 계산하는 myfact() 함수를 재귀적 함수로 구현하시오.
####•[옵션] 참을 수 있는 정도 만큼의 수행시간을 직접 기다려보면서 n을 늘려보도록 합시다


```python
def myfact(n):
    if n==1:
        return 1
    else:
        return n * myfact(n-1)
print myfact(5)
print myfact(20)
```

####myfact 함수는 입력받은 n이 1일 때 1을 반환하고, 그 외에는 반환이 n * myfact(n-1)이다. n==1이 될때까지 계속해서 반복하는 재귀적 함수이다.

####10 이하의 소수를 모두 더하면 2 + 3 + 5 + 7 = 17 이 됩니다.
####이백만(2,000,000) 이하 소수의 합은 얼마입니까?

```python
prime = [2]
n = 1
for x in range (3, 2000001, 2):
    for y in prime:
        if x%y ==0:
            n = 0
            break
    if n == 1:
        prime = prime + [x]
    n = 1
print sum(prime)
            
```

#### prime리스트는 소수를 저장하는 리스트이다. 처음은 2만 들어있다. n은 소수인지 아닌지를 판별해주는 변수로써 소수가 아닐 경우 0으로 된다. 소수는 자신보다 작은 소수로 나누어지면 소수가 아니라는 점을 이용하여 만들어 주었다. 만약 소수로 나누어졌을 경우 n = 0을 넣어주었으며, n ==1이 유지 될 시에 prime리스트에 그 수를 넣어 주었다.

####1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

####1: 1
####3: 1, 3

####10: 1, 2, 5, 10
####15: 1, 3, 5, 15
####21: 1, 3, 7, 21
####28: 1, 2, 4, 7, 14, 28 
####위에서 보듯이, 5개 이상의 약수를 갖는 첫번째 삼각수는 28입니다.
####그러면 500개 이상의 약수를 갖는 가장 작은 삼각수는 얼마입니까?
###(빠른 출력을 위해 30개 이상의 약수를 갖는 가장 작은 삼각수를 출력)

```python
x = 0
count = 0
for i in range(1,100000):
    x = x + i
    for j in range(1, x+1):
        if x%j == 0:
            count = count + 1
    if count >= 30:
        break
    count = 0    
print x
```

####13 → 40 → 20 → 10 → 5 → 16 → 8 → 4 → 2 → 1
####아직 증명은 되지 않았지만, 이런 과정을 거치면 어떤 수로 시작해도 마지막에는 1로 끝나리라 생각됩니다. 
####(역주: 이것은 콜라츠 추측 Collatz Conjecture이라고 하며, 이런 수들을 우박 hailstone sequence라 부르기도 합니다)
####그러면, 백만(1,000,000) 이하의 수로 시작했을 때 1까지 도달하는데 가장 긴 과정을 거치는 숫자는 얼마입니까?
####참고: 계산 과정 도중에는 숫자가 백만을 넘어가도 괜찮습니다.


```python
max_num = 0
count = 0
max_count = 0
for x in range(10, 1000001):
    n = x
    while n != 1:
        if n%2 == 0:
            n = n/2
            count = count + 1
        else:
            n = 3*n + 1
            count = count + 1
    if count > max_count:
        max_count = count
        max_num = x
    count = 0
        
        
print max_num        
```

#### max_num은 계산 과정이 가장 많은 수를 저장하는 변수이다. 판단하는 숫자의 범위를 10부터 1000000까지로 정한 뒤, 문제의 조건에 따라 count의 수를 세준다. count의 값을 배교하여 가장 큰 계산횟수를 가진 수를 max_num에 넣어 준 뒤 출력한다.

####아래와 같은 2 × 2 격자의 왼쪽 위 모서리에서 출발하여 오른쪽 아래 모서리까지 도달하는 길은 모두 6가지가 있습니다 (거슬러 가지는 않기로 합니다).
####그러면 20 × 20 격자에는 모두 몇 개의 경로가 있습니까?


```python
def factorial(n):
    if n == 1 :
        return n
    else:
        return n * factorial(n-1)

print factorial(40)/(factorial(20)*factorial(20))
```

####이 문제는 조합을 이용한 문제이다. 40C20을 계산한 값을 출력해주면 된다. 이를 위해  factoral함수를 재귀함수를 이용해 만들었다.

```python
f = open("www.habitat.co.uk.words.frequency.txt", 'r')
d1 =  pickle.load(f)
f.close()

f = open("na.leagueoflegends.com.frequency.txt", 'r')
d2 =  pickle.load(f)
f.close()

f = open("www.amazon.com.frequency.txt", 'r')
d3=  pickle.load(f)
f.close()

f = open("www.isu.org.frequency.txt", 'r')
d4 =  pickle.load(f)
f.close()

f = open("www.ufc.com.frequency.txt", 'r')
d5 =  pickle.load(f)
f.close()

D = {}
D.update(d1)
D.update(d2)
D.update(d3)
D.update(d4)
D.update(d5)

max1 = ""
max2 = ""
max3 = ""
maxnum1 = 0
maxnum2 = 0
maxnum3 = 0

for x in D:
    if D[x] > maxnum3:
        maxnum3 = D[x]
        max3 = x
    if maxnum3 > maxnum2:
        maxnum2, maxnum3 = maxnum3, maxnum2
        max2, max3 = max3, max2
    if maxnum2 > maxnum1:
        maxnum1, maxnum2 = maxnum2, maxnum1
        max1, max2 = max2, max1

print max1, max2, max3

print D[max1], D[max2], D[max3]
```
